```python
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    pred_adjusted = [val-epsilon if val==1 else val+epsilon for val in y_predicted]
    pred_adjusted = np.array(pred_adjusted)
    return -np.mean(y_true*np.log(pred_adjusted)+(1-y_true)*np.log(1-pred_adjusted))
```

```python
def sigmoid_numpy(X):
    return 1/(1+np.exp(-X))
```

![gradient descent algorithm](/static/gradient_descent.png)

```python
def gradient_descent(age, affordibility, y_true, epochs):
    # w1, w2, bias
    w1 = w2 = 1
    bias = 0

    rate = 0.1
    n = len(age)
    for i in range(epochs):
        # weighted sum
        weighted_sum = w1*age + w2*affordibility+bias
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss(y_true, y_predicted)

        w1d = (1/n)* np.dot(np.transpose(age),(y_predicted - y_true))
        w2d = (1/n) * np.dot(np.transpose(affordibility),(y_predicted-y_true))
        bias_d = np.mean(y_predicted-y_true)

        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        bias = bias - rate * bias_d
        print(f"Epoch: {i}, {w1=}, {w2=}, {bias=}, {loss=}")
    return (w1,w2,bias)

```

### model class

```python
class myNN:
    def __init__(self, initial_w1=1, initial_w2=1, initial_bias=0):
        self.w1 = initial_w1
        self.w2 = initial_w2
        self.bias = initial_bias

    def gradient_descent(self, age, affordibility, y_true, epochs, loss_threshold):
        # w1, w2, bias
        w1,w2,bias = self.w1, self.w2, self.bias

        rate = 0.1
        n = len(age)
        for i in range(epochs):
            # weighted sum
            weighted_sum = w1*age + w2*affordibility+bias
            y_predicted = sigmoid_numpy(weighted_sum)
            loss = log_loss(y_true, y_predicted)

            w1d = (1/n)* np.dot(np.transpose(age),(y_predicted - y_true))
            w2d = (1/n) * np.dot(np.transpose(affordibility),(y_predicted-y_true))
            bias_d = np.mean(y_predicted-y_true)

            w1 = w1 - rate * w1d
            w2 = w2 - rate * w2d
            bias = bias - rate * bias_d
            if i%50==0:
                print(f"Epoch: {i}, {w1=}, {w2=}, {bias=}, {loss=}")
            if loss<=loss_threshold:
                print(f"Epoch: {i}, {w1=}, {w2=}, {bias=}, {loss=}")
                break
            
        return (w1,w2,bias)
    
    def fit(self, X,y, epochs, loss_thresshold=0.4):
        (w1,w2,bias) = self.gradient_descent(X['age'],X['affordibility'],y, epochs, loss_thresshold)
        self.w1,self.w2,self.bias = w1,w2,bias

    def predict(self, X):
        weighted_sum = self.w1*X['age'] + self.w2*X['affordibility']+self.bias
        return 1/(1+np.exp(-weighted_sum))
```

```python
my_nn = myNN()
my_nn.fit(X_train, y_train, 10000,0.4631)
```
