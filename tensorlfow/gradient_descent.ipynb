```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Dropout, Flatten, Input
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("insurance_data.csv")
df['age'] = df['age']/100
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[["age","affordibility"]], df.bought_insurance, test_size=0.2, random_state=41)
X_train.shape
```

```python
model = keras.Sequential([
    Dense(1, input_shape=(2,), activation="sigmoid",kernel_initializer="ones", bias_initializer="zeros")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5000)
```

```python
model.evaluate(X_test, y_test)
```

```python
coef, intercept = model.get_weights()
```

```python
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
def prediction_fucntion(age, affordibility):
    # the constants are taken from gradient descent
    result = coef[0][0]*age+coef[1][0]*affordibility+intercept[0]
    return sigmoid(result)
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    pred_adjusted = [val-epsilon if val==1 else val+epsilon for val in y_predicted]
    pred_adjusted = np.array(pred_adjusted)
    return -np.mean(y_true*np.log(pred_adjusted)+(1-y_true)*np.log(1-pred_adjusted))
```

```python
def sigmoid_numpy(X):
    return 1/(1+np.exp(-X))
```

![gradient descent algorithm](/static/gradient_descent.png)

```python
def gradient_descent(age, affordibility, y_true, epochs):
    # w1, w2, bias
    w1 = w2 = 1
    bias = 0

    rate = 0.1
    n = len(age)
    for i in range(epochs):
        # weighted sum
        weighted_sum = w1*age + w2*affordibility+bias
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss(y_true, y_predicted)

        w1d = (1/n)* np.dot(np.transpose(age),(y_predicted - y_true))
        w2d = (1/n) * np.dot(np.transpose(affordibility),(y_predicted-y_true))
        bias_d = np.mean(y_predicted-y_true)

        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        bias = bias - rate * bias_d
        print(f"Epoch: {i}, {w1=}, {w2=}, {bias=}, {loss=}")
    return (w1,w2,bias)

```
