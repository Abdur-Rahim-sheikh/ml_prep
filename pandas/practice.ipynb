```python
data = {
    "date": ["2025-01-01", "2025-01-02", "2025-01-03", "2025-01-04", "2025-01-05", "2025-01-06"],
    "time_taken": [120, 45, 75, 95, 110, 60],
    "user_id": [101, 102, 103, 104, 101, 102],
    "product": ["Laptop", "Smartphone", "Tablet", "Monitor", "Keyboard", "Mouse"],
    "status": ["Completed", "Pending", "Completed", "Cancelled", "Pending", "Completed"]
}

df = pd.DataFrame(data)
df.head()
```
